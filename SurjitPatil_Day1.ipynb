```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned = df_cleaned.dropna(subset=['title'])
```

```python
df_cleaned['description'] = df_cleaned['description'].fillna('No description available')
```

```python
df_cleaned['age_certification'] = df_cleaned['age_certification'].fillna('NR')
```

```python
import ast
```

```python
df_cleaned['genres'] = df_cleaned['genres'].apply(ast.literal_eval)
```

```python
df_cleaned['production_countries'] = df_cleaned['production_countries'].apply(ast.literal_eval)
```

```python
df_cleaned['seasons'] = df_cleaned['seasons'].fillna(0).astype(int)
```

```python
num_cols = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in num_cols:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())
```

```python
df_cleaned['imdb_id'] = df_cleaned['imdb_id'].fillna('tt0000000')
```
